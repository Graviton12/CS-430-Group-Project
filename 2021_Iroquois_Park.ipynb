## Exploratory Data Analysis for the Iroquois 2021 Data

### 1) Is there any large correlations between different variables of the dataset?

### 2) Do a majority of the invasive species occur frequently? Meaning, is there a major presence of invasive species?


```python
iro_data_2021.columns
```

```python
# Creating histogram for the distribution of invasiveness levels
invasiveness_values = iro_data_2021[['Lonicera japonica', 'Microstegium vimineum', 'Euonymus alatus',
       'Ampelopsis brevipedunculata', 'Morus alba', 'Celastrus orbiculatus',
       'Vinca minor', 'Euonymus fortunei', 'Akebia quinata',
       'Fallopia japonica', 'Achyranthes', 'Hederea helix', 'L. maackii',
       'Privet/ Ligustrum', 'Ailanthus']].values.flatten()

# Define custom bin edges
bin_edges = [0, 1, 2, 3, 4, 5]

# Plotting the histogram
plt.hist(invasiveness_values, bins=bin_edges, color='salmon', edgecolor='maroon')

# Setting the title
plt.title('Distribution of Invasiveness Levels for Entire Dataset in 2021')

# Setting the x-label
plt.xlabel('Invasiveness Level (1 = "Scarce" to 5 = "Many")')

# Aligning the x-ticks to the rightmost edge of each bin
plt.xticks(bin_edges[1:], labels=['1', '2', '3', '4', '5'])

# Setting the y-label
plt.ylabel('Frequency')

# Displaying plot
plt.show()
```

### 3) For all invasive species in the dataset, which invasive species occur most frequently and may pose the greatest threat?

```python
# creatig a copy of the dataset but only focusing on the invasive species rows
invasive_species_dataset = iro_data_2021.iloc[:,3:23]

# adding a row to the dataset which sums the values for each invasise species and sorting it from greatest to smallest
invasive_species_sum = invasive_species_dataset.sum().sort_values(ascending=True)
invasive_species_dataset.loc['sum'] = invasive_species_sum
invasive_species_dataset
```

```python
plt.figure(figsize=(12, 6))

# creating a bar graph to display the invasive level for each invasive species using the sum row
bar_graph = plt.barh(invasive_species_sum.index, invasive_species_sum.values, color='teal')

plt.title('Distribution of Invasiveness Level for Each Invasive Species in 2021')
plt.xlabel('Sum of Invasiveness Level')
plt.ylabel('Invasive Species')

# Add the sum values as labels on the bars
for bar in bar_graph:
    # Getting the sum caue for each bar
    bar_sum_value = bar.get_width()
    # setting the postion for the label for each bar
    plt.text(bar_sum_value + 0.1, bar.get_y() + bar.get_height()/2,
             f'{bar_sum_value: .0f}', va='center')  # The value label

# Display the plot
plt.show()
```

### 4) For all trimble stations, which trimble station has the greatest number of invasive species present

```python
# creating a column invasive_sum where each from is the sum from columns 3-17
iro_data_2021['row_sum'] = iro_data_2021.iloc[:, 3:17].sum(axis=1)
iro_data_2021.head()
```

```python
# filter the data so that only the trimble station's whose coressponding row_sum value is at least 115
iro_top_trimble_stations_2021 = iro_data_2021[iro_data_2021['row_sum'] >= 15]

# creating a histogram to show the occurances of invasive species per trimble station for the top trimble stations
plt.figure(figsize=(9, 6))
# Create a list of positions (0, 1, 2, etc.) for the categorical x-axis
x_positions = list(range(len(iro_top_trimble_stations_2021['Trimble_station'])))

# Create the bar graph using categorical x-axis
bars_2021 = plt.bar(x_positions, iro_top_trimble_stations_2021['row_sum'], color='crimson')

# Adding labels on top of each bar
for bar in bars_2021:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom')

# Adding title
plt.title('Trimble Stations with the most Occurances of Invasive Species in 2021')

# Adding the axes titles
plt.ylabel('Frequency')
plt.xlabel('Trimble Station')
# Set the x-ticks to the station names (categories)
plt.xticks(x_positions, iro_top_trimble_stations_2021['Trimble_station'])

# Display the plot
plt.show()
```

### 5) What are the Top 10 trimble stations and bottom 10 trimble stations in 2021?

```python
#plotting trible station invasive species sum... the top 10, sorting from largest to smallest ... adding a gap between columns
sns.barplot(x='Trimble_station', y='row_sum', data=iro_data_2021.sort_values(by='row_sum', ascending=False).head(10))
#plotting the trimble station invasive species sum bottom 10
sns.barplot(x='Trimble_station', y='row_sum', data=iro_data_2021.sort_values(by='row_sum', ascending=False).tail(10))
plt.title('Top 10 and Bottom 10 Trimble Stations for Invasive Presence')
plt.xlabel('Trimble Station')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Invasive Species Sum')
plt.show()
```
