## Data Preprocessing - 2024 Data

```python
# importing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading dataset
# using skiprows=1 to skip the first row of the dataset
invasive_data_2024=pd.read_excel("/content/drive/MyDrive/CS 430 Group Project/olmstead_dataset.xlsx", sheet_name="2024 Cherokee Invasive Survey", skiprows=1)
```

```python
# viewing the dataset
invasive_data_2024.head(6)
```

```python
# Viewing the names of the columns
invasive_data_2024.columns
```

```python
# looking at the shape of the dataset
invasive_data_2024.shape
```

```python
# checking the datatypes for each column
invasive_data_2024.info()
```

```python
# Viewing the summary statistics for the dataset
invasive_data_2024.describe()
```

### Transforming the Data

```python
# renaming columns
invasive_data_2024 = invasive_data_2024.rename(columns={'Comment':'Trimble_station', 'Calculated using formula in densiometer':'average_canopy_cover', 'Calculated using formula in densiometer.1':'form_densiometer.1'})
```

```python
# checking if columns were renamed
invasive_data_2024.head(5)
```

```python
# Removing the word 'Cherokee' from the Trimble station entries that wya it is left with just its number
invasive_data_2024['Trimble_station'] = invasive_data_2024['Trimble_station'].str.replace('Cherokee', '')

# checking if the entries were changed
invasive_data_2024.head(5)
```

```python
# deleting uneeded columns
invasive_data_2024 = invasive_data_2024.drop(['Unnamed: 27', 'N', 'E', 'S', 'W', 'Canopy gap and/or downed trees present', 'SUM', 'Canopy', 'Understory', 'Dioscorea polystachya'], axis=1)
```

```python
# checking for any missing values
invasive_data_2024.isna().sum()
```

### Handling MIssing Values

```python
#filling out missing values for all columns with 0
invasive_data_2024.fillna(0, inplace=True)
```

```python
# chekcing if values were filled
invasive_data_2024.isna().sum()
```

## Exploratory Data Analysis for the 2024 Data

### 1) Is there any large correlations between different variables of the dataset?

```python
# checking for any correlations between the variables in the dataset
# creating a correlation heat map for all the numerical variables without the Fallopia Japonica species since it has entries of zero.
new_invasive_data_2024 = invasive_data_2024.drop(['Fallopia japonica'], axis=1)
plt.figure(figsize=(12.5, 10))
correlation_matrix = new_invasive_data_2024.select_dtypes(include=[np.number]).corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
```

Based on the heatmap, there does not seem to be any significant correlation between any of the variables aside from form_densiomter.1 and average_canopy_cover which has a strong negative correlation of -0.61

### 2) Do a majority of the invasive species occur frequently? Meaning, is there a major presence of invasive species?


```python
# creating histogram for the distribution of invaseness levels

# focusing on the tree species columns from the dataset
invasiveness_values = invasive_data_2024[['Lonicera japonica', 'Microstegium vimineum',
       'Euonymus alatus', 'Ampelopsis brevipedunculata', 'Morus alba',
       'Celastrus orbiculatus', 'Vinca minor', 'Euonymus fortunei',
       'Akebia quinata', 'Fallopia japonica', 'Achyranthes', 'Hederea helix',
       'L. maackii', 'Privet/ Ligustrum', 'Ailanthus']].values.flatten()

# plotting the histogram
plt.hist(invasiveness_values, bins=5, color='orange', edgecolor='brown')

# setting the title
plt.title('Distribution of Invasiveness Levels for Entire Dataset')
# setting the x label
plt.xlabel('Invasiveness Level (1 = "Scarce" to 5 = "Many")')
plt.xticks(np.arange(0,6))
# setting the y label
plt.ylabel('Frequency')

# displaying plot
plt.show()
```

The histogram shows that for the invasive species in the 2024 Cherokee Invasive Survey dataset as a whole, the majority of the occurances of these invasive speces were within the "scarce" level while the fewest number of occurences were within the "many" level. This implies that for the invasise species in Cherokee park in 2024, the occruances of seeing these invasive species is mostly scarce. This is potienlly a good sign as it shows that a majority of the invasive species do not occur very frequently.

### 3) For all invasive species in the dataset, which invasive species occur most freuquently and may pose the greatest threat?

```python
# creatig a copy of the dataset but only focusing on the invasive species rows
invasive_species_dataset = invasive_data_2024.iloc[:,3:23]
#invasive_species_dataset

# adding a row to the dataet which sums the values for each invasise species and sorting it from greatest to smallest
invasive_species_sum = invasive_species_dataset.sum().sort_values(ascending=True)
invasive_species_dataset.loc['sum'] = invasive_species_sum
invasive_species_dataset
```

```python
plt.figure(figsize=(12, 6))

# creating a bar graph to display the invasive level for each invasive species using the sum row
bar_graph = plt.barh(invasive_species_sum.index, invasive_species_sum.values, color='purple')

plt.title('Distribution of Invasiveness Level for Each Invasive Species')
plt.xlabel('Sum of Invasiveness Level')
plt.ylabel('Invasive Species')

# Add the sum values as labels on the bars
for bar in bar_graph:
    # Getting the sum caue for each bar
    bar_sum_value = bar.get_width()
    # setting the postion for the label for each bar
    plt.text(bar_sum_value + 0.1, bar.get_y() + bar.get_height()/2,
             f'{bar_sum_value: .0f}', va='center')  # The value label

# Display the plot
plt.show()
```

For the 2024 Cherokee Park dataset, based on the bar graph, the Euonymus Fortunei species had the most occurances with a value of 382. Therefore, this species may pose the greatest threat for the park in terms of its level of invaseness. L. maackii had the second highest number of occurances with a value of 210 however this was substantionally smaller that Euonymus Fortunei. Moreover, the Fallopia Japonica species had no occurances. Therefore, this species does not seem to pose a threat at this moment.

### 4) For all trimble station, which trimble station has the greatest number of invasove species present?

```python
# Calculate the sum of all invasive species for each location (row-wise sum)
invasive_data_2024['row_sum'] = invasive_data_2024.drop('Trimble_station', axis=1).sum(axis=1)

# filter the data so that only the trimble station's whose coressponding row_sum value is at least 115
top_trimble_stations = invasive_data_2024[invasive_data_2024['row_sum'] >= 115]

# creating a histogram to show the occurances of invasive species per trimble station for the top trimble stations
plt.figure(figsize=(14, 6))
# Create a list of positions (0, 1, 2, etc.) for the categorical x-axis
x_positions = list(range(len(top_trimble_stations['Trimble_station'])))

# Create the bar graph using categorical x-axis
bars = plt.bar(x_positions, top_trimble_stations['row_sum'], color='darkred')

# Adding labels on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom')

# Adding labels and title (optional)
plt.title('Trimble Stations with the most Occurances of Invasive Species')

# Adding the axes titles
plt.ylabel('Frequency')
plt.xlabel('Trimble Station')
# Set the x-ticks to the station names (categories)
plt.xticks(x_positions, top_trimble_stations['Trimble_station'])

# Display the plot
#plt.tight_layout()
plt.show()
```

The bar graph displasy the top trimble stations with the most occurances of insvaive species. Based on the bar graph for the 2024 Cherokee Park dataset, it is clear that Trimble Station 153 has the greatest number of invasive species with 126 occurances. Thus, this may be an area which will need greater attention. The Trimble stations with the next highest number of invasive species were Trimble station 159 and Timble Station 8 with a value of 121 each.
